# PRACTICING DATA VISUALISATION WITH SEABORN

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
sns.set_style('darkgrid')
```

```python
flowers = sns.load_dataset('iris')
flowers
```

## Let's create some line plots

```python
flowers.plot(title='A Beautiful plot of IRIS', figsize=(20,6));
```

```python
flowers.plot('species', 'sepal_width', figsize=(20,6), c='pink')
```

```python
flowers.plot('species', 'sepal_length', figsize=(20,6), c='red', marker='x')
```

```python
flowers.plot('species', 'petal_width', figsize=(20,6), c='black', marker='d');
```

```python
flowers.plot('species', 'petal_length', figsize=(20,6), c='purple', marker='v')
```

## Bar Plots? yes pleaseee :)

```python
sns.set_style('dark')
```

```python
flowers.plot('species', 'sepal_width', figsize=(20,6), color='pink', kind='bar', width=0.8);
```

```python
flowers.plot('sepal_length', 'sepal_width', figsize=(10,6), color='black', kind='barh', width=0.8);
# this looks so chaotic lol
```

# Let's create  Histogram plots :)

```python
flowers.hist(figsize=(10,6), bins=10);
```

```python
flowers.plot(kind='hist', subplots=True);
flowers.plot(kind='hist', subplots=False, alpha=0.3);
flowers.plot(kind='hist', subplots=False, alpha=0.9);
flowers.plot(kind='hist', subplots=False, alpha=0.1);
```

## Let's do some Scatter Plots

```python
fig,ax= plt.subplots(figsize=(20,6))
ax.scatter(flowers['species'], flowers['sepal_width'], c='grey', marker='d');
```

```python
fig,ax= plt.subplots(figsize=(20,6))
ax.scatter(flowers['species'], flowers['sepal_length'], c='blue', marker='v')
```

```python
fig,ax= plt.subplots(figsize=(20,6))
ax.scatter(flowers['petal_width'], flowers['sepal_width'], c='grey', marker='d')
ax.set(title='Petal Width vs Sepal Width',
      ylabel='Sepal Width',
      xlabel="Pedal Width");
```

```python
fig,ax= plt.subplots(figsize=(20,7))
sns.scatterplot(y=flowers['sepal_length'], x=flowers['sepal_width'], hue=flowers.species, s=100);
plt.title('Comparing Sepal Dimensions');
```

## PLT.FILL_BETWEEN ?

```python
plt.fill_between(flowers['species'], flowers['sepal_width'],flowers['sepal_length'], color='green', alpha=0.4);
```

```python
plt.fill_between(flowers['species'], flowers['petal_width'],flowers['petal_length'], color='yellow', alpha=0.4);
```
